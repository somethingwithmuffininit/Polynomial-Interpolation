```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([1,5,7,8,10])
y = np.array([2, 26, 50, 658, 1018])

# Number of data points
n = len(x)

# Create the Vandermonde matrix
m = np.array([x**i for i in range(n)]).T

# Solve for the polynomial coefficients
coefficients = np.linalg.solve(m, y)

# Generate x values for plotting
x2 = np.linspace(min(x), max(x), 500)

# Evaluate the polynomial using the coefficients
function = np.polyval(coefficients, x2)

plt.figure(figsize=(8, 6))
plt.plot(x2, function, color='blue')
plt.axhline(0, color='black', linewidth=0.8, linestyle='--')
plt.axvline(0, color='black', linewidth=0.8, linestyle='--')
plt.title("Polynomial Plot")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()

plt.plot(x,y)

plt.grid()
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt

# Given data points
x = np.array([23521, 55253, 72523, 8523, 12350])
y = np.array([25862,5568565 ,685658688 ,958685688, 1687706856800])

# Number of data points
n = len(x)

# Create the Vandermonde matrix
m = np.array([x**i for i in range(n)]).T

# Solve for the polynomial coefficients
coefficients = np.linalg.solve(m, y)

# Generate x values for plotting
x2 = np.linspace(min(x), max(x), 5050)

# Set the threshold
threshold = 0.00000001

# Initialize an empty list to store the components
components = []

# Iterate over the vector and add components to the list until they are small enough
flag = 0
for val in coefficients:
    if abs(val) >= threshold:
        flag = 0
    else:
        flag +=1
        if flag > 1:
            components = components[:-1]
            break
    components.append(val)




# Convert the list back to a NumPy array
components = np.array(components)





# Evaluate the polynomial using the coefficients
function = np.polyval(components, x2)

# Plot the polynomial
plt.figure(figsize=(8, 6))
plt.plot(x2, function, color='blue', label="Interpolating Polynomial")
plt.scatter(x, y, color='red', label="Data Points")  # Add data points
plt.axhline(0, color='black', linewidth=0.8, linestyle='--')
plt.axvline(0, color='black', linewidth=0.8, linestyle='--')
plt.title("Polynomial Interpolation")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid()
plt.show()





```
